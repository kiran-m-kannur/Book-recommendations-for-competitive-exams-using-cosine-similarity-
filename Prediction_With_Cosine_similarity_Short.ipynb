```python
#lib
import numpy as np 
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
columns = ['Title','Authors','Genre']
```

```python
def combine_features(data):
  features = []
  for i in range(0, data.shape[0]):
    features.append(data['Title'][i] + ' '+data['Author'][i]+' '+data['Genre'][i])
  return features
```

```python
df['combined_features']=combine_features(df)
df.head(50)
```

```python
cm = CountVectorizer().fit_transform(df['combined_features'])#matrix converstion
cs = cosine_similarity(cm)#cosine similarity
print(cs)
```

```python
Title = df['Title'][43]
Title
```

```python
book_id = df[df.Title == Title]['Book_id'].values[0]-1
book_id
```

```python

scores = list(enumerate(cs[book_id]))
print(scores)
```

```python
sorted_scores = sorted(scores,key=lambda x:x[1],reverse=True)
sorted_scores = sorted_scores[1:]
sorted_scores
```

```python
j=0
print('5 Most recommended books to '+Title+' are:\n')
for item in sorted_scores:
  book_title = df[df.Book_id == item[0]]['Title'].values[0]
  print(j+1,book_title)
  j=j+1
  if j>=5:
    break
```
